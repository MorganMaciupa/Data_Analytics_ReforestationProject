```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('default')


```

```python

forest_dataframe = pd.read_csv("https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2021/2021-04-06/forest.csv")
forest_area_dataframe = pd.read_csv("https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2021/2021-04-06/forest_area.csv")
```

```python
forest_area_dataframe
```

```python
forest_dataframe
```

```python
forest_data_complete = pd.merge(forest_area_dataframe, forest_dataframe, how="outer", on=["entity", "entity"])
forest_data_complete
```

```python
forest_data_complete = forest_data_complete.fillna(0)
forest_data_complete
```

```python
africa_df = forest_area_dataframe.loc[forest_area_dataframe["entity"] == "Africa", :]

africa_df
```

```python

africa_df.plot.line(x='year', y='forest_area', color="red", label="forest area",
                    title= " Deforestation in Africa");
plt.xlabel("Years");
plt.ylabel("Forest area Mha");

plt.savefig("Images/Deforestation in Africa.png")
```

```python
europe_df = forest_area_dataframe.loc[forest_area_dataframe["entity"] == "Europe", :]
europe_df=europe_df.drop('code',1);
europe_df
```

```python
europe_df.plot.line(x='year', y='forest_area', color="green", label="forest area",
                    title= "Deforestation in Europe");
plt.xlabel("Years");
plt.ylabel("Forest area Mha ");

plt.savefig("Images/Deforestation in Europe.png")
```

```python
asia_df = forest_area_dataframe.loc[forest_area_dataframe["entity"] == "Asia", :]
asia_df=asia_df.drop('code',1);
asia_df
```

```python
asia_df.plot.line(x='year', y='forest_area', color="blue", label="Forest Area",
                    title= "Deforestation in Asia");
plt.xlabel("Years");
plt.ylabel("Forest Area Mha");

plt.savefig("Images/Deforestation in Asia.png")
```

```python
oceania_df= forest_area_dataframe.loc[forest_area_dataframe["entity"]=="Oceania",:]

oceania_df
```

```python
oceania_df.plot.line(x='year', y='forest_area', color="pink", label="Forest Area",
                    title= "Deforestation in Oceania");
plt.xlabel("Years");
plt.ylabel("Forest Area Mha");

plt.savefig("Images/Deforestation in Oceania.png")
```

```python

australia_df = forest_area_dataframe.loc[forest_area_dataframe["entity"] == "Australia", :]

australia_df
```

```python
australia_df.plot.line(x='year', y='forest_area', color="black", label="Forest Area",
                    title= " Deforestation in Australia");
plt.xlabel("Years");
plt.ylabel("Forest Area Mha");

plt.savefig("Images/Deforestation in Australia.png")
```

```python
australia_oceania_df = pd.merge(australia_df, oceania_df, how="outer",on="year")
australia_oceania_df =australia_oceania_df.drop(columns=['code_x','code_y'])
sum_column = australia_oceania_df["forest_area_x"]+australia_oceania_df["forest_area_y"]
australia_oceania_df["forest_area"]= sum_column
australia_oceania_df.head()
```

```python
australia_oceania_df.plot.line(x='year', y='forest_area', color="purple", label="Forest Area",
                    title= "Deforestation in Australia/ Oceania");
plt.xlabel("Years");
plt.ylabel("Forest Area Mha");

plt.savefig("Images/Deforestation in Australia & Oceania.png")
```

```python
south_america_df = forest_area_dataframe.loc[forest_area_dataframe["entity"] == "South America", :]
south_america_df=south_america_df.drop('code',1);

south_america_df
```

```python
south_america_df.plot.line(x='year', y='forest_area', color="gray", label="Forest Area",
                    title= " Deforestation in South America");
plt.xlabel("Years");
plt.ylabel("Forest Area Mha");

plt.savefig("Images/Deforestation in Soputh America.png")
```

```python
united_state_df = forest_area_dataframe.loc[forest_area_dataframe["entity"] == "United States", :]

united_state_df
```

```python
canada_df = forest_area_dataframe.loc[forest_area_dataframe["entity"] == "Canada", :]

canada_df
```

```python
north_america_df = pd.merge(united_state_df, canada_df, how="outer",on="year")
north_america_df =north_america_df.drop(columns=['code_x','code_y'])
sum_column = north_america_df["forest_area_x"]+north_america_df["forest_area_y"]
north_america_df["forest_area"]= sum_column
north_america_df.head(40)
```

```python
north_america_df.plot.line(x='year', y='forest_area', color="brown", label="Forest Area",
                    title= "Deforestation in North America");


plt.xlabel("Years");
plt.ylabel("Forest Area Mha");

plt.savefig("Images/Deforestation in North America.png")
```

```python


plt.plot(asia_df['year'], asia_df['forest_area'])
plt.plot(africa_df['year'], africa_df['forest_area'])
plt.plot(europe_df['year'], europe_df['forest_area'])
plt.plot(australia_oceania_df['year'], australia_oceania_df['forest_area'])
plt.plot(south_america_df['year'], south_america_df['forest_area']);
plt.plot(north_america_df['year'],north_america_df['forest_area'])

plt.xlabel("Years");
plt.ylabel("Forest Area");
plt.title("World Deforestation");

labels = ['Asia','Africa','Australia/Oceania','Europe','South America','North America']
plt.legend(labels,loc="best");

plt.savefig("Images/Deforestation in World .png")
```

```python
fig, axs = plt.subplots(2, 3, sharex='all')

fig.suptitle("Deforestation by Continent")
years =[1990,2000,2010,2020]


axs[0, 0].plot(asia_df['year'], asia_df['forest_area'])
axs[0, 0].set_title("Asia")

axs[0,1].set_xticks(years)

axs[1, 0].plot(africa_df['year'], africa_df['forest_area'])
axs[1, 0].set_title("Africa")

axs[0, 1].plot(europe_df['year'], europe_df['forest_area'])
axs[0, 1].set_title("Europe")

axs[1, 1].plot(australia_oceania_df['year'], australia_oceania_df['forest_area'])
axs[1, 1].set_title("Australia/Oceania")

axs[0, 2].plot(south_america_df['year'], south_america_df['forest_area'])
axs[0, 2].set_title("South America")

axs[1, 2].plot(north_america_df['year'], north_america_df['forest_area'])
axs[1, 2].set_title("North America")

axs[0,0].tick_params(labelbottom=True)
axs[0,1].tick_params(labelbottom=True)
axs[0,2].tick_params(labelbottom=True)

plt.savefig("Images/Deforestation global.png")

fig.tight_layout()
```
